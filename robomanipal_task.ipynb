```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
cols = ['WindDir9am', 'WindDir3pm', 'WindGustDir']

for col in cols:
    data_copy[col] = le.fit_transform(data_copy[col])
```

```python
from sklearn.preprocessing import StandardScaler

num_cols = [
    'MinTemp', 'MaxTemp', 'Rainfall', 'WindGustSpeed',
    'WindSpeed9am', 'WindSpeed3pm',
    'Humidity9am', 'Humidity3pm',
    'Pressure9am', 'Pressure3pm',
    'Temp9am', 'Temp3pm'
]

scaler = StandardScaler()
data_copy[num_cols] = scaler.fit_transform(data_copy[num_cols])
```

```python
data_copy.head(7)
```

```python
# Example one-hot encoding
data_copy = pd.get_dummies(data_copy, columns=['Location'], drop_first=True)
data_copy['RainTomorrow'] = data_copy['RainTomorrow'].map({'No': 0, 'Yes': 1})

```

```python
data_copy['Date'] = pd.to_datetime(data_copy['Date'])
data_copy['IsWeekend'] = data_copy['DayOfWeek'].isin([5,6]).astype(int)
```

```python
def get_season(month):
    if month in [12, 1, 2]: return 'Summer'
    elif month in [3, 4, 5]: return 'Autumn'
    elif month in [6, 7, 8]: return 'Winter'
    else: return 'Spring'
data_copy['Season'] = data_copy['Month'].apply(get_season)
data_copy = pd.get_dummies(data_copy, columns=['Season'], drop_first=True)
```

```python
data_copy['TempRange'] = data_copy['MaxTemp'] - data_copy['MinTemp']
data_copy['TempDiff9am3pm'] = data_copy['Temp3pm'] - data_copy['Temp9am']
```

```python
data_copy['HumidityDiff'] = data_copy['Humidity3pm'] - data_copy['Humidity9am']
data_copy['WindSpeedDiff'] = data_copy['WindSpeed3pm'] - data_copy['WindSpeed9am']
```

```python
data_copy = data_copy.sort_values(['Location', 'Date'])
data_copy['Rainfall_3dayAvg'] = data_copy.groupby('Location')['Rainfall'].transform(lambda x: x.rolling(3, min_periods=1).mean())
data_copy['Rainfall_Lag1'] = data_copy.groupby('Location')['Rainfall'].shift(1)
data_copy['Temp3pm_Lag1'] = data_copy.groupby('Location')['Temp3pm'].shift(1)
```
