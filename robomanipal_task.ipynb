```python
import numpy as np

def remove_outliers_iqr(df, columns):
    for col in columns:
        Q1 = df[col].quantile(0.25)
        Q3 = df[col].quantile(0.75)
        IQR = Q3 - Q1
        lower = Q1 - 1.5 * IQR
        upper = Q3 + 1.5 * IQR
        # Filter within bounds
        df = df[(df[col] >= lower) & (df[col] <= upper)]
    return df

numeric_cols = data_copy.select_dtypes(include=['float64', 'int64']).columns
data_no_outliers = remove_outliers_iqr(data_copy, numeric_cols)
```
